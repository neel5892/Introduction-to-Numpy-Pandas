```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(10)

student_data = {
    "Student_ID": np.arange(1, 51),
    "Screen_Time_Hours": np.random.uniform(2, 10, 50).round(2),
    "Study_Hours": np.random.uniform(1, 6, 50).round(2),
    "Sleep_Hours": np.random.uniform(4, 9, 50).round(2),
    "Social_Media_Usage_Hours": np.random.uniform(0.5, 5, 50).round(2),
    "Battery_Drain_Percent": np.random.randint(25, 100, 50)
}

df = pd.DataFrame(student_data)
df.head()
```

```python
df.to_csv("student_smartphone_usage.csv", index=False)
print("Dataset saved successfully.")
```

```python
data = pd.read_csv("student_smartphone_usage.csv")
data.head()
```

CELL 7: NumPy Array Operations

```python
screen_time = data["Screen_Time_Hours"].values

print("Average Screen Time:", np.mean(screen_time))
print("Maximum Screen Time:", np.max(screen_time))
print("Minimum Screen Time:", np.min(screen_time))
print("Standard Deviation:", np.std(screen_time))
```

CELL 8: Filtering Data Using Pandas

```python
high_screen_users = data[data["Screen_Time_Hours"] > 7]
high_screen_users.head()
```

CELL 9: Students with Low Sleep

```python
low_sleep = data[data["Sleep_Hours"] < 6]
low_sleep[["Student_ID", "Sleep_Hours", "Screen_Time_Hours"]]
```

CELL 10: Correlation Analysis

```python
data.corr()
```

CELL 11: Feature Engineering

```python
data["Productivity_Score"] = (data["Study_Hours"] * 2) - data["Screen_Time_Hours"]
data.head()
```

CELL 12: Sorting Data

```python
data.sort_values(by="Productivity_Score", ascending=False).head()
```

CELL 13: Basic Visualization

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(data["Screen_Time_Hours"], bins=10)
plt.xlabel("Screen Time (Hours)")
plt.ylabel("Number of Students")
plt.title("Distribution of Screen Time")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix
corr_matrix = data.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Student Smartphone Usage Dataset")
plt.show()
```
